# How does the distribution of features evolve when we go through the timebags?

## Some loading and prep code

```python
# First some imports
import numpy as np
import matplotlib as mpl
import seaborn as sb
%matplotlib inline

import settings as st
from linguistics.treetagger import get_tagger
from linguistics.wn import lemmatize as relemmatize
from datainterface.redistools import RedisReader
import datainterface.picklesaver as ps
from util.generic import ProgressInfo
tagger = get_tagger()
```

```python
# Load the clusters, the features, and initialize the results
clusters = RedisReader(st.redis_mt_clusters_filtered_pref)
features = dict((fsrc, dict((ftype, ps.load(fdetails['file'].format('all')))
                            for ftype, fdetails in fsrc_details.iteritems()))
                for fsrc, fsrc_details in st.mt_analysis_features.iteritems())
empirical_features = dict((fsrc, dict((ftype, [])
                                      for ftype in fsrc_values.iterkeys()))
                          for fsrc, fsrc_values in features.iteritems())
```

```python
# A useful function for later on
def get_filtered_relems(tbg, ref_ids):
    """List the relemmatized words (once per occurrence) appearing in `tbg`,
    excluding words from quotes in `ref_ids`
    
    To build the distribution of word features for a timebag, we want the
    list of all words appearing in the timebag, once per occurrence. And to
    do that for only new quotes appeared in a later timebag, we can exclude
    quotes whose ids are in `ref_ids` (typically the set of quote ids from
    the first timebag of the cluster).
    
    Quotes are first lemmatized with the TreeTagger lemmatizer, then
    relemmatized as in the mining analysis, with the WordNet morphy
    method.
    
    Parameters
    ----------
    tbg : TimeBag
        The timebag to work on.
    ref_ids : set
        The set of ids from which to exclude quotes in `tbg`.
    
    Returns
    -------
    list
        The list of relemmatized words.
    
    """
    
    filtered_relems = []

    for string, freq, strid in zip(tbg.strings, tbg.tot_freqs, tbg.ids):
        # If our current string is in the set of banned strings,
        # jump to next one
        if strid in ref_ids:
            continue

        relems = [relemmatize(lem) for lem in tagger.Lemmatize(string)]
        filtered_relems.extend(relems * freq)

    return filtered_relems
```

## The analysis itself

We want to look at the distribution of features in *new* quotes. So for each cluster, we look at the first timebag, and take it as a reference. Then for each later timebag, we compute the distribution of features for the quotes that appear in that timebag but weren't in the first one (i.e. new quotes from timebag 0). This gives us the distribution of features on the novel quotes.

```python
# Duration of a timebag, in days
timebag_size = 2.0
# Minimum number of timebags a cluster must have to be included in the analysis
min_number_of_timebags = 10
```

```python
try:
    # See if we already computed this
    empirical_features = ps.load('data/analyses/empirical_features.pickle')
    print 'Loaded from pickle.'
except IOError:
    # A progress indicator
    progress = ProgressInfo(len(clusters), 10, 'features')
    
    # For each cluster
    for c in clusters.itervalues():
        progress.next_step()
        
        # Build timebags, skip if we don't have enough bags
        tbgs = c.build_timebags(timebag_size)
        if len(tbgs) < min_number_of_timebags:
            continue
        
        ref_ids = set(tbgs[0].ids)
        for i, tbg in enumerate(tbgs):
            # For the first timebag, don't filter (i.e. the reference set is empty)
            filtered_relems = get_filtered_relems(tbg, [] if i == 0 else ref_ids)
            
            # For each feature
            for fsrc, fsrc_values in features.iteritems():
                for ftype, feature in fsrc_values.iteritems():
            
                    # Get the feature values for the extracted words
                    tbg_empirical_features = [feature[relem] for relem in filtered_relems if relem in feature]
                    try:
                        # Either add it to the existing list
                        empirical_features[fsrc][ftype][i].extend(tbg_empirical_features)
                    except IndexError:
                        # Or create the list for this timebag if necessary
                        empirical_features[fsrc][ftype].append(tbg_empirical_features)
                        # Then check that the item we just created was indeed for the right timebag.
                        empirical_features[fsrc][ftype][i]
            
    # Convert to numpy arrays
    for fsrc, fsrc_values in features.iteritems():
        for ftype in fsrc_values.iterkeys():
            for i, all_tbg_empirical_features in enumerate(empirical_features[fsrc][ftype]):
                empirical_features[fsrc][ftype][i] = np.array(all_tbg_empirical_features)
    
    # Save to disk
    ps.save(empirical_features, 'data/analyses/empirical_features.pickle')
```

## A quick check

Just to have a look, here we count the average number of *new* quotes that appear in the first ten timebags of each cluster (*new* according to the definition above). It looks pretty stationary.

```python
try:
    # See if we already computed this
    np_new_quotes_counts = ps.load('data/analyses/new_quotes_counts.pickle')
    print 'Loaded from pickle.'
except IOError:
    n_clusters_kept = 0
    new_quotes_counts = []
    progress = ProgressInfo(len(clusters), 10, 'clusters')
    
    for c in clusters.itervalues():
        progress.next_step()
        
        tbgs = c.build_timebags(timebag_size)
        if len(tbgs) < min_number_of_timebags:
            continue
        
        n_clusters_kept += 1
        new_quotes_counts.append([len(set(tbgs[i].ids).difference(set(tbgs[0].ids))) for i in range(min_number_of_timebags)])
    
    print 'Kept {} clusters.'.format(n_clusters_kept)
    
    np_new_quotes_counts = np.array(new_quotes_counts)
    
    # Save to disk
    ps.save(np_new_quotes_counts, 'data/analyses/new_quotes_counts.pickle')

mpl.pyplot.plot(np_new_quotes_counts.mean(0))
```

## Plot the results

We now plot the distribution of features for novel quotes, for each feature, for a number of timebags. And see if it evolves. Unfortunatly it doesn't seem to.

```python
# How many bins we want
nbins = 10

# How many timebags to plot
n_timebags_shown = 10

# Step between two plotted timebags
n_timebags_step = 2

# For each feature
for fsrc, fsrc_empirical_values in empirical_features.iteritems():
    for ftype, tbgs_empirical_values in fsrc_empirical_values.iteritems():
        
        # Check if we're using a log-scale
        if st.mt_analysis_features[fsrc][ftype]['log']:
            log = True
            logtext = ' (log)'
        else:
            log = False
            logtext = ''
        
        # Create axes
        fig, ax = mpl.pyplot.subplots(1, 1, figsize=(20, 4))
        fig.text(0.5, 1, '{} {}'.format(fsrc, ftype.replace('_', ' ')) + logtext)
        
        # Convert values
        vals = []
        for lvals in tbgs_empirical_values[:n_timebags_shown * n_timebags_step:n_timebags_step]:
            
            if log:
                vals.append(np.log(lvals))
            else:
                vals.append(lvals)
            
        names = ['tbg {}'.format(i) for i in range(0, n_timebags_shown * n_timebags_step, n_timebags_step)]
        
        # We want the same bandwidth in all violins for a given feature
        # n**(-1./(d+4)) where n is the number of data points,
        # and d the number of dimensions (Scott's rule)
        bw = len(vals[0]) ** (- 1. / (1 + 4))
        sb.violinplot(vals, names=names, bw=bw, ax=ax, log=log)
```

## Ideas

* Maybe the distributions don't change because new quotes with base distribution keep being injected. What happens if we restrict to quotes in further timebags that are in the sphere around timebag 0, or around the previous timebag?
* Now if the total timebags don't evolve over time (because of reinjection, or other transformations), we can do the above point restricting to quotes in the sphere around quotes we selected in the previous timebag (i.e. those quotes that were themselves in the sphere from the previous timebag, etc. recursively).

At any rate, it's clear that there's no global effect (which makes sense since the whole distribution is probably stationary), no evolution of the whole over time. Which is a problem for Sperber. Which means, **we need to characterize quote injection.**
