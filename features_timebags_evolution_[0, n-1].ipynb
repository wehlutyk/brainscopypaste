# How does the distribution of features evolve when we go through the timebags?

## Some loading and prep code

```python
# First some imports
import numpy as np
import matplotlib as mpl
import seaborn as sb
%matplotlib inline

import settings as st
from linguistics.treetagger import get_tagger
from linguistics.wn import lemmatize as relemmatize
from datainterface.redistools import RedisReader
import datainterface.picklesaver as ps
from util.generic import ProgressInfo
tagger = get_tagger()
```

```python
# Load the clusters, the features, and initialize the results
clusters = RedisReader(st.redis_mt_clusters_filtered_pref)
features = dict((fsrc, dict((ftype, ps.load(fdetails['file'].format('all')))
                            for ftype, fdetails in fsrc_details.iteritems()))
                for fsrc, fsrc_details in st.mt_analysis_features.iteritems())
empirical_features = dict((fsrc, dict((ftype, [])
                                      for ftype in fsrc_values.iterkeys()))
                          for fsrc, fsrc_values in features.iteritems())
```

```python
from linguistics.distance import hamming_word
    
def is_in_sphere(ref_strings, string):
    """Whether or not `string` is in the hamming_word-sphere at
    distance 1 surrounding the `ref_strings`."""
    
    for ref_str in ref_strings:
        if hamming_word(ref_str, string) == 1:
            return True
    return False


def get_filtered_relems_and_strings(tbg, ref_strings):
    """List the relemmatized words (once per occurrence) appearing in `tbg`,
    excluding words from strings in `ref_strings`, as well as the new strings
    taken into account.
    
    To build the distribution of word features for a timebag, we want the
    list of all words appearing in the timebag, once per occurrence. And to
    do that for only new quotes compared to an earlier set of quotes, we can
    exclude quotes that are already in `ref_strings` (typically the first
    timebag, or the output of previous filterings of timebags to restrict to
    quotes produced by substitution) and that aren't in the immediate sphere
    around strings in `ref_tbg`.
    
    Quotes are first lemmatized with the TreeTagger lemmatizer, then
    relemmatized as in the mining analysis, with the WordNet morphy
    method.
    
    Parameters
    ----------
    tbg : TimeBag
        The timebag to work on.
    ref_strings : list of strings
        The reference string from which to exclude; if None, then ignored.
    
    Returns
    -------
    list
        The list of relemmatized words.
    set
        The set of new strings considered to the product of this iteration.
    
    """
    
    filtered_relems = []
    new_strings = set([])

    for string, freq in zip(tbg.strings, tbg.tot_freqs):
        # If our current string is in the set of banned strings,
        # jump to next one
        if ref_strings is not None and string in ref_strings:
            continue
        
        # If our current string is not in the sphere around ref_strings,
        # jump to next one
        if ref_strings is not None and not is_in_sphere(ref_strings, string):
            continue

        new_strings.add(string)
        relems = [relemmatize(lem) for lem in tagger.Lemmatize(string)]
        filtered_relems.extend(relems * freq)

    return filtered_relems, new_strings
```

## The analysis itself

We want to look at the distribution of features in *new* quotes produced by susbtitution (i.e. not those reinjected from outside as part of the continuous process). Here, for each timebag in a cluster, we look at the take the quotes that haven't appeared in any of the previous timebags at all. Then on each of these sets of quotes, we compute the distribution of features. This gives us the distribution of features on the novel quotes.

```python
# Duration of a timebag, in days
timebag_size = 2.0
# Minimum number of timebags a cluster must have to be included in the analysis
min_number_of_timebags = 10
```

```python
try:
    # See if we already computed this
    empirical_features = ps.load('data/analyses/empirical_features_[0, n-1].pickle')
    np_new_quotes_counts = ps.load('data/analyses/new_quotes_counts_[0, n-1].pickle')
    print 'Loaded from pickle.'
except IOError:
    n_clusters_kept = 0
    new_quotes_counts = []
    
    # A progress indicator
    progress = ProgressInfo(len(clusters), 10, 'clusters')
    
    # For each cluster
    for c in clusters.itervalues():
        progress.next_step()
        
        # Build timebags, skip if we don't have enough bags
        tbgs = c.build_timebags(timebag_size)
        if len(tbgs) < min_number_of_timebags:
            continue
        
        n_clusters_kept += 1
        tbgs_n_new_quotes = []
        ref_strings = set([])
        for i, tbg in enumerate(tbgs):
            filtered_relems, new_strings = get_filtered_relems_and_strings(tbg, None if i == 0 else ref_strings)
            
            # Update the set of reference strings to represent all timebags up to now, for next iteration
            ref_strings.update(tbg.strings)
            
            # Update number of new strings
            if 0 < i and i < min_number_of_timebags:
                tbgs_n_new_quotes.append(len(new_strings))
            
            # For each feature
            for fsrc, fsrc_values in features.iteritems():
                for ftype, feature in fsrc_values.iteritems():
            
                    # Get the feature values for the extracted words
                    tbg_empirical_features = [feature[relem] for relem in filtered_relems if relem in feature]
                    try:
                        # Either add it to the existing list
                        empirical_features[fsrc][ftype][i].extend(tbg_empirical_features)
                    except IndexError:
                        # Or create the list for this timebag if necessary
                        empirical_features[fsrc][ftype].append(tbg_empirical_features)
                        # Then check that the item we just created was indeed for the right timebag.
                        empirical_features[fsrc][ftype][i]
                        
        new_quotes_counts.append(np.array(tbgs_n_new_quotes))
    
    # Convert to numpy arrays
    print 'Kept {} clusters.'.format(n_clusters_kept)
    np_new_quotes_counts = np.array(new_quotes_counts)
    for fsrc, fsrc_values in features.iteritems():
        for ftype in fsrc_values.iterkeys():
            for i, all_tbg_empirical_features in enumerate(empirical_features[fsrc][ftype]):
                empirical_features[fsrc][ftype][i] = np.array(all_tbg_empirical_features)
    
    # Save to disk
    ps.save(empirical_features, 'data/analyses/empirical_features_[0, n-1].pickle')
    ps.save(np_new_quotes_counts, 'data/analyses/new_quotes_counts_[0, n-1].pickle')
```

```python
x = np.arange(1, min_number_of_timebags)
y = np_new_quotes_counts.mean(0)
ci = 1.96 * np_new_quotes_counts.std(0) / np.sqrt(len(np_new_quotes_counts[:, 0]) - 1)
mpl.pyplot.plot(x, np_new_quotes_counts.mean(0))
mpl.pyplot.fill_between(x, y - ci, y + ci, alpha=0.2, color='grey')
```

## Plot the results

We now plot the distribution of features for novel quotes, for each feature, for a number of timebags. And see if it evolves.

```python
# How many bins we want
nbins = 10

# How many timebags to plot
n_timebags_shown = 10

# Step between two plotted timebags
n_timebags_step = 2

# For each feature
for fsrc, fsrc_empirical_values in empirical_features.iteritems():
    for ftype, tbgs_empirical_values in fsrc_empirical_values.iteritems():
        
        # Check if we're using a log-scale
        if st.mt_analysis_features[fsrc][ftype]['log']:
            log = True
            logtext = ' (log)'
        else:
            log = False
            logtext = ''
        
        # Create axes
        fig1, ax1 = mpl.pyplot.subplots(1, 1, figsize=(20, 4))
        fig1.text(0.5, 1, '{} {}'.format(fsrc, ftype.replace('_', ' ')) + logtext)
        fig2, axs2 = mpl.pyplot.subplots(1, n_timebags_shown, figsize=(20, 3), sharex=True)
        
        # Convert values
        vals = []
        for lvals in tbgs_empirical_values[:n_timebags_shown * n_timebags_step:n_timebags_step]:
            
            if log:
                vals.append(np.log(lvals))
            else:
                vals.append(lvals)
            
        names = ['tbg {}'.format(i) for i in range(0, n_timebags_shown * n_timebags_step, n_timebags_step)]
        
        # We want the same bandwidth in all violins for a given feature
        # n**(-1./(d+4)) where n is the number of data points,
        # and d the number of dimensions (Scott's rule)
        bw = len(vals[0]) ** (- 1. / (1 + 4))
        sb.violinplot(vals, names=names, bw=bw, ax=ax1, log=log)
        
        # Add histograms
        for i in range(n_timebags_shown):
            if i == 0:
                bins = axs2[i].hist(vals[i], bins=nbins, log=log, normed=True, color="#6495ED")[1]
            else:
                axs2[i].hist(vals[i], bins=bins, log=log, normed=True, color="#6495ED")
            axs2[i].set_xlabel('tbg {} ({} pts)'.format(i * n_timebags_step, len(vals[i])))
            axs2[i].set_xlim([bins.min(), bins.max()])
            axs2[i].set_xticklabels(axs2[i].get_xticks(), rotation=60)
```
