```python
# Translating internal feature names to public names
tr_map = {'aoa_Kuperman': 'Age of acquisition',
          'cmu_MNphonemes': r'$\left< \right.$ \# phonemes $\left. \right>$',
          'cmu_MNsyllables': r'$\left< \right.$ \# syllables $\left. \right>$',
          'fa_BCs': 'FA log-betweenness',
          'fa_CCs': 'FA log-clustering',
          'fa_degrees': 'FA log-degree',
          'fa_PR_scores': 'FA log-pagerank',
          'mt_frequencies': 'Log-frequency',
          'mt_start_frequencies': 'Log-frequency in start quotes',
          'wn_BCs': 'WN log-betweenness',
          'wn_CCs': 'WN log-clustering',
          'wn_degrees': 'WN log-degree',
          'wn_MNSyns': r'Log $\left< \right.$ \# synonyms $\left. \right>$',
          'wn_NSigns': r'Log \# meanings',
          'wn_PR_scores': 'WN log-pagerank'}

back_tr_map = dict([(v, k) for k, v in tr_map.iteritems()])

def _translate(labels, tr):
    if isinstance(labels, list):
        return [tr[l] for l in labels]
    else:
        return tr[labels]

def translate(labels):
    """Translate a list of internal labels to human-readable ones."""
    return _translate(labels, tr_map)

def back_translate(labels):
    """Translate a list of human-readable labels to internal ones."""
    return _translate(labels, back_tr_map)
```

```python
# Build susceptibilities and filter out zeros
for a in sa.analyses:
    try:
        a.f_susceptibilities
    except AttributeError:
        a.build_susceptibilities()
        # The try-except here lets us run this cell several times without overwriting
        # the original unfiltered susceptibilities each time.
        a.f_susceptibilities_orig = a.f_susceptibilities
    
    a.f_susceptibilities = a.f_susceptibilities_orig#[np.nonzero(a.f_susceptibilities_orig[:, 1])]
    # Remove MNSyns outlier
    #if a.feature.fullname == 'wn MNSyns all':
    #    a.f_susceptibilities = \
    #            a.f_susceptibilities_orig[np.nonzero(a.f_susceptibilities_orig[:, 1])][:-1]

# Set names, order and colorify analyses
for a in sa.analyses:
    a.feature.shortfullname = a.feature.fullname.replace(' ', '_').rstrip('_all')
    a.feature.publicname = translate(a.feature.shortfullname)
analyses = sorted(sa.analyses, key=lambda a: included_features.index(a.feature.shortfullname))
palette = sb.color_palette('husl', len(included_features))
palette_dict = dict(zip([a.feature.publicname for a in analyses], palette))

# Convert to DataFrame
sdata = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * 
                                                 len(a.f_susceptibilities) for a in analyses]),
                      'feature value': np.concatenate([a.f_susceptibilities[:, 0] for a in analyses]),
                      'substitution susceptibility': np.concatenate([a.f_susceptibilities[:, 1] 
                                                                     for a in analyses])})
```

```python
# Don't close the figure after drawing
%config InlineBackend.close_figures=False

nbins_suscepts = 5
n_pts_threshold = 15

bins_cache = []

# Our plotting function for a given feature
def plot_susceptibility(xvalues, svalues, **kwargs):
    color = kwargs.get('color', 'blue')
    
    #if is_twelfth_int(xvalues):
    #    bins = np.linspace(int(np.floor(xvalues.iloc[0])) - 0.5, int(np.ceil(xvalues.iloc[-1])) + 0.5,
    #                       (int(np.ceil(xvalues.iloc[-1])) - int(np.floor(xvalues.iloc[0])) + 2) / 2)
    #else:
    #    bins = np.linspace(xvalues.iloc[0], xvalues.iloc[-1], nbins)
    bins = np.linspace(xvalues.iloc[0], xvalues.iloc[-1], nbins_suscepts + 1)
    bins_cache.append(bins)
    xs = (bins[1:] + bins[:-1]) / 2
    ys = np.zeros_like(xs)
    cs = np.zeros_like(xs)
    for i in range(len(bins) - 1):
        idx = np.where((bins[i] <= xvalues) * (xvalues < bins[i + 1]))[0]
        if len(idx) > n_pts_threshold:
            yss = svalues.iloc[idx]
            ys[i] = yss.mean()
            cs[i] = 1.96 * yss.std() / np.sqrt(len(idx) - 1)
        else:
            ys[i] = np.nan
            cs[i] = np.nan
    
    #plt.hlines(0, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    #plt.hlines(1, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.plot(xs, ys, '-', lw=2, color=color)
    plt.fill_between(xs, ys - cs, ys + cs, color=sb.desaturate(color, 0.2), alpha=0.2)
    plt.xlim(xs[np.isfinite(ys)][0], xs[np.isfinite(ys)][-1])
    ylim = plt.ylim()
    #plt.ylim(max([0, ylim[0]]), min([1, ylim[1]]))
    plt.ylim(0, .5)

# Plot!
sg = sb.FacetGrid(sdata, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1.5, size=3, sharex=False, sharey=False, hue='feature', palette=palette_dict)
sg.map(plot_susceptibility, 'feature value', 'substitution susceptibility')
sg.set_titles('{col_name}')
# Remove the last axis and recenter the lower row
#sg.axes[1, 2].set_axis_off()
#pts02 = sg.axes[2].get_position().get_points()
#pts01 = sg.axes[1].get_position().get_points()
#trans = Affine2D().translate((pts02[1, 0] - pts01[1, 0]) / 2, 0)
#box10 = sg.axes[3].get_position()
#box11 = sg.axes[4].get_position()
sg.axes[2].set_xlabel('')
#sg.axes[3].set_position(TransformedBbox(box10, trans))
#sg.axes[4].set_position(TransformedBbox(box11, trans))

# Save bins
feature_bins = {}
for i in range(5):
    feature_bins[back_translate(sg.axes[i].get_title())] = bins_cache[i]

# Figure is saved later on, once we've added the heatmap
```

### Synonyms ~ Phonemes susceptibility heatmap

```python
analysis_names = [a.feature.fullname for a in sa.analyses]
a_syns = sa.analyses[analysis_names.index('wn MNSyns all')]
a_phonemes = sa.analyses[analysis_names.index('cmu MNphonemes all')]
a_freq = sa.analyses[analysis_names.index('mt frequencies all')]
a_cc = sa.analyses[analysis_names.index('fa CCs all')]

a_syns.build_susceptibilities()
a_phonemes.build_susceptibilities()

ws_syns = a_syns.w_susceptibilities
wf_syns = {}
wf_syns_full = a_syns.feature.data
wf_freq_full = a_freq.feature.data
wf_cc_full = a_cc.feature.data
ws_phonemes = a_phonemes.w_susceptibilities
wf_phonemes = {}
wf_phonemes_full = a_phonemes.feature.data
w_syns_phonemes = set(ws_syns.keys()).intersection(ws_phonemes.keys())

n_diff = 0
for w in w_syns_phonemes:
    n_diff += ws_syns[w] != ws_phonemes[w]
    try:
        feature_syns = a_syns.feature(w)
        feature_phonemes = a_phonemes.feature(w)
    except KeyError:
        continue
    wf_syns[w] = feature_syns
    wf_phonemes[w] = feature_phonemes

# Note, since MNSyns lemmatizes and MNphonemes does not, there are different susceptibilities values
# for some words. In what follows, we take the value from MNSyns (i.e. lemmatizing)
print("{} words (out of {}) with non-matching susceptibilities".format(n_diff,
                                                                       len(w_syns_phonemes)))
```

```python
# Build 2D data

nbins_syns = 5
nbins_phonemes = 5
n_pts_threshold = -1

#range_syns = feature_ranges['wn_MNSyns']
#range_phonemes = feature_ranges['cmu_MNphonemes']
#bins_syns = np.linspace(range_syns[0], range_syns[1], nbins_syns + 1)
#bins_phonemes = np.linspace(range_phonemes[0], range_phonemes[1], nbins_phonemes + 1)
bins_syns = feature_bins['wn_MNSyns']
bins_phonemes = feature_bins['cmu_MNphonemes']
suscepts2d = np.zeros((nbins_syns, nbins_phonemes))
freqs2d = np.zeros_like(suscepts2d)
n2d = np.zeros_like(suscepts2d)
s_all = []

for i in range(nbins_syns):

    bin_syns = bins_syns[i:i+2]
    w_bin_syns = set([w for w, f in wf_syns.iteritems() if bin_syns[0] <= f < bin_syns[1]])
    if i == nbins_syns:
        # On last bin, include right limit
        w_bin_syns.update([w for w, f in wf_syns.iteritems() if f == bin_syns[1]])

    for j in range(nbins_phonemes):

        bin_phonemes = bins_phonemes[j:j+2]
        w_bin_phonemes = set([w for w, f in wf_phonemes.iteritems()
                              if bin_phonemes[0] <= f < bin_phonemes[1]])
        if j == nbins_phonemes:
            # On last bin, include right limit
            w_bin_phonemes.update([w for w, f in wf_phonemes.iteritems()
                                   if f == bin_phonemes[1]])
        
        w_bin_syns_phonemes = w_bin_syns.intersection(w_bin_phonemes)
        s_bin_syns_phonemes = [ws_syns[w] for w in w_bin_syns_phonemes]
        s_all.extend(s_bin_syns_phonemes)
        freq_bin_syns_phonemes = [a_freq.feature(w) for w in w_bin_syns_phonemes]
        n2d[i, j] = len(s_bin_syns_phonemes)
        freqs2d[i, j] = np.mean(freq_bin_syns_phonemes)
        suscepts2d[i, j] = np.mean(s_bin_syns_phonemes) \
            if len(s_bin_syns_phonemes) > n_pts_threshold else np.nan

print n2d
print suscepts2d
print freqs2d
print "Average susceptibility over all words in heatmap: {}".format(np.mean(s_all))
```

### Words involved in substitutions

```python
from IPython.utils import coloransi
tc = coloransi.TermColors()

def red(s):
    return tc.Red + s + tc.Normal

def blue(s):
    return tc.Blue + s + tc.Normal

def green(s):
    return tc.Green + s + tc.Normal

def black(s):
    return tc.Black + s + tc.Normal

def print_subs(subs, indent):
    for s in subs:
        tokens = tagger.Tokenize(s.mother)
        text = ' '.join(tokens[:s.idx]) + ' '
        text += red('{{{} => {}}}'.format(s.word1, s.word2))
        text += ' ' + ' '.join(tokens[s.idx+1:])
        freq = "{:.2f}".format(wf_freq_full[s.lem2]) if s.lem2 in wf_freq_full else '-'
        cc = "{:.2f}".format(wf_cc_full[s.lem2]) if s.lem2 in wf_cc_full else '-'
        text += blue(" [synonyms {}, phonemes {}, freq {}, cc {}, susceptibility {}]".format(
            wf_syns_full.get(s.lem2, '-'), wf_phonemes_full.get(s.lem2, '-'), 
            freq, cc, ws_syns.get(s.lem2, '-')))
        print indent * ' ' + text
```

```python
#detail_bin_indices = [[2, 0], [1, 3]]
detail_bin_indices = [(i, j) for i in range(4) for j in range(4)]
detail_bins = [[bins_syns[i:i+2], bins_phonemes[j:j+2]] for i, j in detail_bin_indices]

for bin_syns, bin_phonemes in detail_bins:
    w_bin_syns = set([w for w, f in wf_syns.iteritems() if bin_syns[0] <= f < bin_syns[1]])
    w_bin_phonemes = set([w for w, f in wf_phonemes.iteritems()
                          if bin_phonemes[0] <= f < bin_phonemes[1]])
    w_bin_syns_phonemes = w_bin_syns.intersection(w_bin_phonemes)
    
    # TITLE
    title = "#synonyms {} --- #phonemes {}".format(bin_syns, bin_phonemes)
    print("=" * len(title))
    print(title)
    print
    
    # FOR EACH WORD
    indent = 2
    for w in w_bin_syns_phonemes:
        freq = "{:.2f}".format(wf_freq_full[w]) if w in wf_freq_full else '-'
        cc = "{:.2f}".format(wf_cc_full[w]) if w in wf_cc_full else '-'
        print (indent - 1) * '-' + " " + red(w) + \
            " (synonyms {}, phonemes {}, freq {}, cc {}, susceptibility {})".format(
                wf_syns[w], wf_phonemes[w], freq, cc, ws_syns[w]),
        
        # Find back the substitutions
        subs = [s for s in a_syns.data if s.lem1 == w or s.word1 == w]
        if len(subs) == 0:
            print '[nosubs]'
            print
            continue

        print
        print
        print_subs(subs, indent + 2)
        print
        
    print
```

## Feature variations on substitution

```python
nbins_variations = 5
n_pts_threshold = 10

# Build variations
for a in analyses:
    if a.feature.fullname == 'cmu MNphonemes all':
        a.set_bins(nbins_variations, truncmax=25)
    else:
        a.set_bins(nbins_variations)
    a.build_h0()
    # Force rebuild of variations in case our n_pts_threshold has changed
    try:
        del a.daughter_d
    except AttributeError:
        pass
    a.build_variations(n_pts_threshold)

# Convert to DataFrame
vdata_binned = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.bin_middles) 
                                                        for a in analyses]),
                             'disappearing bin middles': np.concatenate([a.bin_middles
                                                                         for a in analyses]),
                             'appearing bin averages': np.concatenate([a.daughter_d
                                                                       for a in analyses]),
                             'appearing H0': np.concatenate([a.daughter_d_h0
                                                             for a in analyses]),
                             'appearing H0n': np.concatenate([a.daughter_d_h0_n
                                                              for a in analyses]),
                             'ci': np.concatenate([a.daughter_d_ci for a in analyses])})
vdata_raw = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.l2_f_mothers)
                                                     for a in analyses]),
                          'disappearing raw': np.concatenate([a.l2_f_mothers for a in analyses]),
                          'appearing raw': np.concatenate([a.l2_f_daughters for a in analyses]),
                          'appearing H0': np.concatenate([np.ones_like(a.l2_f_daughters) *
                                                          a.feature.values.mean() for a in analyses])})
```

```python
with_h0n = False

def plot_variations(xvalues, avalues, h0s, h0ns, cis, **kwargs):
    color = kwargs.get('color', 'blue')
    
    plt.plot(xvalues, avalues, '-', lw=2, color=color, label=r'$\nu_{\phi}$')
    plt.plot(xvalues, h0s, linestyle='--', color=sb.desaturate(color, 0.2), 
             label=r'$\nu_{\phi}^0$')
    if with_h0n:
        plt.plot(xvalues, h0ns, linestyle='-.', color=sb.desaturate(color, 0.2), 
                 label=r'$\nu_{\phi}^{00}$')
    plt.plot(xvalues, xvalues, linestyle='dotted', color=sb.desaturate(color, 0.2), label=r'$y = x$')
    plt.fill_between(xvalues.values, avalues.values - cis.values, avalues.values + cis.values,
                     color=sb.desaturate(color, 0.2), alpha=0.2)
    xmin, xmax = xvalues.iloc[0], xvalues.iloc[-1]
    plt.xlim(xmin, xmax)
    #plt.text(xmin + (xmax - xmin) * 0.05, h0s.iloc[-1], r'$\mathcal{H}_0$', ha='left', va='center',
    #         bbox=dict(facecolor=sb.desaturate('#EAEAF2', 0.2), alpha=0.7))
    # Rescale on Y if we're touching h0, making sure the h0 line is visible
    ymin, ymax = plt.ylim()
    if ymin == h0s.min():
        ymin -= (ymax - ymin) / 10
    elif ymax == h0s.max():
        ymax += (ymax - ymin) / 10
    plt.ylim(ymin, ymax)

def plot_facetgrid():
    vg = sb.FacetGrid(vdata_binned, col='feature', col_wrap=3, 
                      col_order=translate(included_features),
                      aspect=1.5, size=3, sharex=False, sharey=False, hue='feature',
                      palette=palette_dict)#,
                      #legend=True)
    vg.map(plot_variations, 'disappearing bin middles', 'appearing bin averages', 
           'appearing H0', 'appearing H0n', 'ci')
    vg.set_titles('{col_name}')
    vg.set_xlabels(r'$f \left( \right.$ disappearing word $\left. \right)$')
    vg.set_ylabels(r'$f \left( \right.$ appearing word $\left. \right)$')
    for ax in vg.axes.ravel():
        legend = ax.legend(frameon=True, loc='upper left')
        if not legend:
            # Skip if nothing was plotted on these axes
            continue
        frame = legend.get_frame()
        frame.set_facecolor('#f2f2f2')
        frame.set_edgecolor('#000000')
    # Remove the last axis and recenter the lower row
    #vg.axes[1, 2].set_axis_off()
    pts02 = vg.axes[2].get_position().get_points()
    pts01 = vg.axes[1].get_position().get_points()
    trans = Affine2D().translate((pts02[1, 0] - pts01[1, 0]) / 2, 0)
    box10 = vg.axes[3].get_position()
    box11 = vg.axes[4].get_position()
    vg.axes[2].set_xlabel('')
    vg.axes[3].set_position(TransformedBbox(box10, trans))
    vg.axes[4].set_position(TransformedBbox(box11, trans))
    
    if with_h0n:
        filename = 'data/figures_paper/feature_variations-binned-with-H00.png'
    else:
        filename = 'data/figures_paper/feature_variations-binned.png'
    vg.fig.savefig(filename, bbox_inches='tight', dpi=300)

plot_facetgrid()
with_h0n = True
plot_facetgrid()
```
