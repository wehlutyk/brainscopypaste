# How do features vary on substitution? What susceptibility to substitution do they induce?

## Data-loading and formatting

See below for graphs.

```python
# Initial imports
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D, TransformedBbox
import numpy as np
import seaborn as sb
import pandas as pd

from analyze.args import AnalysisArgs
from analyze.substitutions import SubstitutionsAnalyzer
from util.generic import is_twelfth_int
```

```python
included_features = ['aoa_Kuperman',
                     'mt_frequencies',
                     'fa_CCs',
                     'cmu_MNphonemes',
                     'wn_MNSyns']

# Build the analysis classes
aa = AnalysisArgs({'ff': 'ff',
                   'model': 'tbgs',
                   'POS': 'all',
                   'substrings': False,
                   'timebag_size': 1.0,
                   'features': included_features,
                   'positions': False,
                   'paths': False,
                   'save': False,
                   'overwrite': False,
                   'show': True})
sa = SubstitutionsAnalyzer(aa)

print 'Loaded the {} substitutions that have been mined.'.format(len(sa.substitutions))
```

## Susceptibility to substitution for all the features

*feature value* ~ *susceptibility value*

**All plots are with zero values removed.** This consists in looking at the susceptibilities *for words that have been substituted at least once*.

```python
# Translating internal feature names to public names
tr_map = {'aoa_Kuperman': 'Age of acquisition',
          'cmu_MNphonemes': r'$\left< \right.$ \# phonemes $\left. \right>$',
          'cmu_MNsyllables': r'$\left< \right.$ \# syllables $\left. \right>$',
          'fa_BCs': 'FA log-betweenness',
          'fa_CCs': 'FA log-clustering',
          'fa_degrees': 'FA log-degree',
          'fa_PR_scores': 'FA log-pagerank',
          'mt_frequencies': 'Log-frequency',
          'mt_start_frequencies': 'Log-frequency in start quotes',
          'wn_BCs': 'WN log-betweenness',
          'wn_CCs': 'WN log-clustering',
          'wn_degrees': 'WN log-degree',
          'wn_MNSyns': r'Log $\left< \right.$ \# synonyms $\left. \right>$',
          'wn_NSigns': r'Log \# meanings',
          'wn_PR_scores': 'WN log-pagerank'}

def translate(labels):
    """Translate a list of internal labels to human-readable ones."""
    if isinstance(labels, list):
        return [tr_map[l] for l in labels]
    else:
        return tr_map[labels]
```

```python
# Build susceptibilities and filter out zeros
for a in sa.analyses:
    try:
        a.f_susceptibilities
    except AttributeError:
        a.build_susceptibilities()
        # The try-except here lets us run this cell several times without overwriting
        # the original unfiltered susceptibilities each time.
        a.f_susceptibilities_orig = a.f_susceptibilities
    
    a.f_susceptibilities = a.f_susceptibilities_orig[np.nonzero(a.f_susceptibilities_orig[:, 1])]

# Set names, order and colorify analyses
for a in sa.analyses:
    a.feature.shortfullname = a.feature.fullname.replace(' ', '_').rstrip('_all')
    a.feature.publicname = translate(a.feature.shortfullname)
analyses = sorted(sa.analyses, key=lambda a: included_features.index(a.feature.shortfullname))
palette = sb.color_palette('husl', len(included_features))
palette_dict = dict(zip([a.feature.publicname for a in analyses], palette))

# Convert to DataFrame
sdata = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.f_susceptibilities) for a in analyses]),
                      'feature value': np.concatenate([a.f_susceptibilities[:, 0] for a in analyses]),
                      'substitution susceptibility': np.concatenate([a.f_susceptibilities[:, 1] for a in analyses])})
```

```python
n_pts_threshold = 1

# Our plotting function for a given feature
def plot_susceptibility(xvalues, svalues, **kwargs):
    color = kwargs.get('color', 'blue')
    
    if is_twelfth_int(xvalues):
        bins = np.arange(int(np.floor(xvalues.iloc[0])), int(np.ceil(xvalues.iloc[-1])) + 2) - 0.5
    else:
        bins = np.linspace(xvalues.iloc[0], xvalues.iloc[-1], 20)
    xs = (bins[1:] + bins[:-1]) / 2
    ys = np.zeros_like(xs)
    cs = np.zeros_like(xs)
    for i in range(len(bins) - 1):
        idx = np.where((bins[i] <= xvalues) * (xvalues < bins[i + 1]))[0]
        if len(idx) > n_pts_threshold:
            yss = svalues.iloc[idx]
            ys[i] = yss.mean()
            cs[i] = 1.96 * yss.std() / np.sqrt(len(idx) - 1)
        else:
            ys[i] = np.nan
            cs[i] = np.nan
    
    #plt.hlines(0, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    #plt.hlines(1, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.plot(xs, ys, '-', lw=2, color=color)
    plt.fill_between(xs, ys - cs, ys + cs, color=sb.desaturate(color, 0.2), alpha=0.2)
    plt.xlim(xs[0], xs[-1])
    plt.ylim(0, 1)

# Plot!
sg = sb.FacetGrid(sdata, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1.5, size=3, sharex=False, hue='feature', palette=palette_dict)
sg.map(plot_susceptibility, 'feature value', 'substitution susceptibility')
sg.set_titles('{col_name}')
# Remove the last axis and recenter the lower row
sg.axes[1, 2].set_axis_off()
pts12 = sg.axes[1, 2].get_position().get_points()
pts11 = sg.axes[1, 1].get_position().get_points()
trans = Affine2D().translate((pts12[1, 0] - pts11[1, 0]) / 2, 0)
box10 = sg.axes[1, 0].get_position()
box11 = sg.axes[1, 1].get_position()
sg.axes[1, 0].set_position(TransformedBbox(box10, trans))
sg.axes[1, 1].set_position(TransformedBbox(box11, trans))
#sg.fig.savefig('paper-figures/feature_susceptibilities.png', bbox_inches='tight', dpi=300)
```

## Feature variations on substitution

The first 4 rows are the network properties (betweenness, clustering, pagerank, and degrees), computed on the Free Association network (graphs to the left) and the WordNet network (graphs to the right).

The last 4 rows are the remaining features ("usual suspects").

```python
# Build variations
for a in analyses:
    a.build_h0()
    # Force rebuild of variations in case our n_pts_threshold has changed
    try:
        del a.daughter_d
    except AttributeError:
        pass
    a.build_variations(n_pts_threshold)

# Convert to DataFrame
vdata_binned = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.bin_middles) for a in analyses]),
                             'disappearing bin middles': np.concatenate([a.bin_middles for a in analyses]),
                             'appearing bin averages': np.concatenate([a.daughter_d for a in analyses]),
                             'appearing H0': np.concatenate([np.ones_like(a.bin_middles) * a.feature.values.mean() for a in analyses]),
                             'ci': np.concatenate([a.daughter_d_ci for a in analyses])})
vdata_raw = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.l2_f_mothers) for a in analyses]),
                          'disappearing raw': np.concatenate([a.l2_f_mothers for a in analyses]),
                          'appearing raw': np.concatenate([a.l2_f_daughters for a in analyses]),
                          'appearing H0': np.concatenate([np.ones_like(a.l2_f_daughters) * a.feature.values.mean() for a in analyses])})
```

```python
def plot_variations(xvalues, avalues, h0s, cis, **kwargs):
    color = kwargs.get('color', 'blue')
    
    plt.plot(xvalues, avalues, '-', lw=2, color=color)
    plt.plot(xvalues, h0s, linestyle='dashed', color=sb.desaturate(color, 0.2), label=r'$\mathcal{H}_0$')
    plt.fill_between(xvalues.values, avalues.values - cis.values, avalues.values + cis.values,
                     color=sb.desaturate(color, 0.2), alpha=0.2)
    xmin, xmax = xvalues.iloc[0], xvalues.iloc[-1]
    plt.xlim(xmin, xmax)
    #plt.text(xmin + (xmax - xmin) * 0.05, h0s.iloc[-1], r'$\mathcal{H}_0$', ha='left', va='center',
    #         bbox=dict(facecolor=sb.desaturate('#EAEAF2', 0.2), alpha=0.7))
    # Rescale on Y if we're touching h0, making sure the h0 line is visible
    ymin, ymax = plt.ylim()
    if ymin == h0s.min():
        ymin -= (ymax - ymin) / 10
    elif ymax == h0s.max():
        ymax += (ymax - ymin) / 10
    plt.ylim(ymin, ymax)

# Plot!
vg = sb.FacetGrid(vdata_binned, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1.5, size=3, sharex=False, sharey=False, hue='feature', palette=palette_dict,
                  legend=True)
vg.map(plot_variations, 'disappearing bin middles', 'appearing bin averages', 'appearing H0', 'ci')
vg.set_titles('{col_name}')
vg.set_xlabels(r'$f \left( \right.$ disappearing word $\left. \right)$')
vg.set_ylabels(r'$f \left( \right.$ appearing word $\left. \right)$')
for ax in vg.axes.ravel():
    legend = ax.legend(frameon=True, loc='upper left')
    if not legend:
        # Skip if nothing was plotted on these axes
        continue
    frame = legend.get_frame()
    frame.set_facecolor('#f2f2f2')
    frame.set_edgecolor('#000000')
# Remove the last axis and recenter the lower row
vg.axes[1, 2].set_axis_off()
pts12 = vg.axes[1, 2].get_position().get_points()
pts11 = vg.axes[1, 1].get_position().get_points()
trans = Affine2D().translate((pts12[1, 0] - pts11[1, 0]) / 2, 0)
box10 = vg.axes[1, 0].get_position()
box11 = vg.axes[1, 1].get_position()
vg.axes[1, 0].set_position(TransformedBbox(box10, trans))
vg.axes[1, 1].set_position(TransformedBbox(box11, trans))
#vg.fig.savefig('paper-figures/feature_variations-binned.png', bbox_inches='tight', dpi=300)
```

```python
def kdeplot(xvalues, yvalues, h0s, **kwargs):
    color = kwargs.get('color', 'blue')
    sb.kdeplot(pd.DataFrame({'X': xvalues, 'Y': yvalues}), shade=True)
    xmin, xmax = plt.xlim()
    plt.plot(xvalues, h0s, linestyle='dashed', color=sb.desaturate(color, 0.2))
    plt.xlim(xmin, xmax)

vg = sb.FacetGrid(vdata_raw, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1, size=4, sharex=False, sharey=False)
vg.map(kdeplot, 'disappearing raw', 'appearing raw', 'appearing H0')
vg.set_titles('{col_name}')
vg.set_xlabels(r'$f \left( \right.$ disappearing word $\left. \right)$')
vg.set_ylabels(r'$f \left( \right.$ appearing word $\left. \right)$')
# Remove the last axis and recenter the lower row
from matplotlib.transforms import Affine2D, TransformedBbox
vg.axes[1, 2].set_axis_off()
pts12 = vg.axes[1, 2].get_position().get_points()
pts11 = vg.axes[1, 1].get_position().get_points()
trans = Affine2D().translate((pts12[1, 0] - pts11[1, 0]) / 2, 0)
box10 = vg.axes[1, 0].get_position()
box11 = vg.axes[1, 1].get_position()
vg.axes[1, 0].set_position(TransformedBbox(box10, trans))
vg.axes[1, 1].set_position(TransformedBbox(box11, trans))
```
