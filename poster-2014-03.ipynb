```python
# Initial imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from analyze.args import AnalysisArgs
from analyze.substitutions import SubstitutionsAnalyzer
```

```python
# Build the analysis classes
aa = AnalysisArgs({'ff': 'filtered',
                   'model': 'tbgs',
                   'POS': 'all',
                   'substrings': False,
                   'timebag_size': 5.0,
                   'features': ['cmu_MNphonemes',
                                #'cmu_MNsyllables',
                                'aoa',
                                'fa', 
                                'wn',
                                'mt'],
                   'positions': False,
                   'paths': False,
                   'save': False,
                   'overwrite': False,
                   'show': True})
sa = SubstitutionsAnalyzer(aa)
```

```python
# How many substitutions do we have?
len(sa.substitutions)
```

```python
for i, a in enumerate(sa.analyses):
    print i, a.feature.data_src, a.feature.ftype
```

```python
a_freq = sa.analyses[5]  # MT Frequencies

a_freq.build_h0()
a_freq.build_variations()

fig_freq_var, ax = plt.subplots()
ax.plot(a_freq.bin_middles, a_freq.v_d_h0, 'r', label='$H_0$')
ax.plot(a_freq.bin_middles, a_freq.v_d_h0_n, 'c', label='$H_{0,n$}')
ax.fill_between(a_freq.bin_middles, a_freq.v_d - a_freq.v_d_std, a_freq.v_d + a_freq.v_d_std, color='grey', alpha=0.2)
ax.plot(a_freq.bin_middles, a_freq.v_d, linewidth=2, label='$<\log(f(daughter)) - \log(f(mother))>$')
ax.set_xlabel('$\log(f(mother))$')
ax.set_ylabel('$\log(f(daughter))$')
ax.set_title('Variation of word log-frequencies')
ax.legend(loc='best')
```

```python
fig_freq_sus, ax = plt.subplots()

a_freq.build_susceptibilities()

# Averages in bins
bins = np.linspace(a_freq.f_susceptibilities[0, 0], a_freq.f_susceptibilities[-1, 0], 20)
xs = (bins[1:] + bins[:-1]) / 2
ys = np.zeros_like(xs)
cs = np.zeros_like(xs)
for i in range(len(bins) - 1):
    idx = np.where((bins[i] <= a_freq.f_susceptibilities[:, 0]) * (a_freq.f_susceptibilities[:, 0] < bins[i + 1]))[0]
    yss = a_freq.f_susceptibilities[idx, 1]
    ys[i] = yss.mean()
    cs[i] = 1.96 * yss.std() / np.sqrt(len(idx) - 1)

ax.plot(xs, ys, 'b-', lw=2)
ax.fill_between(xs, ys - cs, ys + cs, color='grey', alpha=0.2)
ax.set_xlabel('$\log(f(mother))$')
ax.set_ylabel('Susceptibility')
ax.set_title('Substitution susceptibility per word log-frequency')
```

```python
# Save figures for MT frequencies
fig_freq_sus.savefig('freq_substitutability.pdf', dpi=150, format='pdf', transparent=True)
fig_freq_var.savefig('freq_variations.pdf', dpi=150, format='pdf', transparent=True)
```
