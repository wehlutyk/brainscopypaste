# Feature correlations

Exploring how features we computed are correlated.

## Load and preformat the data, define utility functions

```python
# Imports
import numpy as np

import datainterface.picklesaver as ps
import settings as st
```

```python
# Load feature data
included_features = ['aoa_Kuperman',
                     'cmu_MNphonemes', 'cmu_MNsyllables',
                     'fa_BCs', 'fa_CCs', 'fa_PR_scores', 'fa_degrees',
                     'mt_frequencies',
                     'wn_NSigns']
features = {}

for fsrc, feature in st.mt_analysis_features.iteritems():
    for ftype, finfo in feature.iteritems():
        fname = fsrc  + '_' + ftype
        if fname not in included_features:
            continue
        
        filename = finfo['file'].format('all')
        features[fname] = ps.load(filename)

# Check we used all the specified included features
assert len(features) == len(included_features)
```

```python
# Get lists of common words between all pairs of features
common_words = {}
for fsrc1, f1 in features.iteritems():
    for fsrc2, f2 in features.iteritems():
        if fsrc1 == fsrc2:
            continue
        cw = set(f1.keys()).intersection(f2.keys())
        if fsrc1 not in common_words or fsrc2 not in common_words:
            common_words[fsrc1] = {}
            common_words[fsrc2] = {}
        if fsrc2 not in common_words[fsrc1] or fsrc1 not in common_words[fsrc2]:
            common_words[fsrc1][fsrc2] = cw
            common_words[fsrc2][fsrc1] = cw
            print '{} words for {} -- {}'.format(len(cw), fsrc1, fsrc2)
```

### Pretty-plotting and label translation functions

```python
# Imports
from __future__ import division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot
from matplotlib.patches import Ellipse
%matplotlib inline

# Our pretty-plotting function
def pretty_plot_correlations(correlations, labels, size_scaling=1.0,
                             cor_fontsize=30.0, label_fontsize=35.0):
    """Pretty-plot a correlation matrix.
    
    Parameters
    ----------
    correlations : np.array
        Sqare matrix containing the correlation values to plot;
        should be symmetric and have ones on the diagonal.
    labels : list of strings
        Labels corresponding to the variables whose correlations are
        to be plotted; the order is the same as for `correlations`.
    size_scaling : float, optional
        Scaling factor for the whole figure; for instance `2.0` would
        create a figure twice as big (for better resolution when saving);
        defaults to 1.0.
    cor_fontsize : float, optional
        Font size for the correlation values; defaults to 30.0.
    label_fontsize : float, optional
        Font size for the labels; defaults to 35.0.
    
    Returns
    -------
    Figure
        The figure created.
    
    """
    
    # Some layout definitions
    ellipse_lwidth = 1.5
    sep_crop = 0.2
    
    # Precompute useful values
    n_features = correlations.shape[0]
    lims = np.linspace(0, 1, n_features + 1)
    step = 1 / n_features
    xs = lims[:-1] + step / 2
    esize = 1.0 * step
    
    # Create our figure
    fig, ax = mpl.pyplot.subplots(figsize=(4 * size_scaling, 4 * size_scaling))
    
    # For all couples
    for i, f1_cor in enumerate(correlations):
        for j, cor in enumerate(f1_cor):
            # Compute coordinates
            x = xs[j]
            y = 1 - xs[i]
            
            if i < j:
                # Plot an Ellipse
                color = mpl.cm.bwr((1 - cor) / 2)
                ax.add_patch(Ellipse([x, y], esize, esize * (1 - abs(cor)),
                             angle=np.sign(cor) * 45, ls='solid', lw=ellipse_lwidth,
                             ec='black', fc=color, alpha=0.8))
            
            elif i > j:
                # Or show the correlation value
                ax.text(x, y, "{:.2f}".format(cor).rstrip("0"),
                        fontsize=cor_fontsize,
                        horizontalalignment='center',
                        verticalalignment='center')
    
    # Next, add the line separators and labels
    for i in range(n_features):
        # Line separator
        ax.plot([(i + sep_crop) * step, (i + 1 - sep_crop) * step],
                [1 - (i + 1 - sep_crop) * step, 1 - (i + sep_crop) * step],
                'k-', lw=ellipse_lwidth)
        
        # Horizontal label
        ax.text(- step / 2, 1 - step / 2 - i * step,
                labels[i], fontsize=label_fontsize,
                horizontalalignment='right', verticalalignment='center')
        
        # Vertical label
        ax.text(step / 2 + i * step, 1 + step / 2,
                labels[i], fontsize=label_fontsize, rotation='vertical',
                horizontalalignment='center', verticalalignment='bottom')
    
    # Finally, set the axis limits and hide the axis themselves
    ax.set_xlim(- 0.1, 1.1)
    ax.set_ylim(- 0.1, 1.1)
    ax.axis('off')
    
    return fig
```

```python
tr_map = {'aoa_Kuperman': 'Age of acquisition',
          'cmu_MNphonemes': r'$\left< \right.$ \# phonemes $\left. \right>$',
          'cmu_MNsyllables': r'$\left< \right.$ \# syllables $\left. \right>$',
          'fa_BCs': 'FA betweenness',
          'fa_CCs': 'FA clustering',
          'fa_degrees': 'FA degree',
          'fa_PR_scores': 'FA pagerank',
          'mt_frequencies': 'Frequency',
          'mt_start_frequencies': 'Frequency in start quotes',
          'wn_BCs': 'WN betweenness',
          'wn_CCs': 'WN clustering',
          'wn_degrees': 'WN degree',
          'wn_MNSyns': r'$\left< \right.$ \# synonyms $\left. \right>$',
          'wn_NSigns': r'\# meanings',
          'wn_PR_scores': 'WN pagerank'}

def translate(labels):
    """Translate a list of internal labels to human-readable ones."""
    return [tr_map[l] for l in labels]
```

### Compute the pearson correlation matrix and pretty-print

```python
## Compute correlation matrix

# Initialize
n = len(features)
labels = []
f_coords = {}
pearson_correlations = np.zeros((n, n))

# Fill in the correlations
for i, (fname, f) in enumerate(sorted(features.iteritems())):
    f_coords[fname] = i
    labels.append(fname)

for i, (fname1, f1) in enumerate(sorted(features.iteritems())):
    for j, (fname2, f2) in enumerate(sorted(features.iteritems())):
        if fname1 == fname2:
            continue
            
        # Get common values
        f1_values = np.array([f1[w] for w in common_words[fname1][fname2]])
        f2_values = np.array([f2[w] for w in common_words[fname1][fname2]])
        
        # Center the vectors
        c_f1_values = f1_values - f1_values.mean()
        c_f2_values = f2_values - f2_values.mean()
        
        # Compute covariance
        pearson_correlations[i, j] = (c_f1_values.dot(c_f2_values) /
                                      np.sqrt(c_f1_values.dot(c_f1_values) * c_f2_values.dot(c_f2_values)))
```

```python
fig = pretty_plot_correlations(pearson_correlations, translate(labels), size_scaling=1.8, cor_fontsize=15, label_fontsize=20)
fig.canvas.print_figure('paper-figures/feature_correlations-filter1.png', dpi=300)
```

```python
fig2 = pretty_plot_correlations(pearson_correlations2, translate(labels2), size_scaling=1.3, cor_fontsize=15, label_fontsize=20)
fig2.canvas.print_figure('paper-figures/feature_correlations-filter2.png', dpi=300)
```
