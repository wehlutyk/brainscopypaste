# Comparison of high index nodes between WordNet and Free Association

Are high index nodes more or less the same words?

## Data loading

```python
from datainterface import picklesaver as ps
wn = {'degrees': ps.load('data/WordNet/wordnet_degrees_all.pickle'),
      'CCs': ps.load('data/WordNet/wordnet_CCs_all.pickle'),
      'BCs': ps.load('data/WordNet/wordnet_BCs_all.pickle'),
      'PR_scores': ps.load('data/WordNet/wordnet_PR_scores_all.pickle')}
fa = {'degrees': ps.load('data/FreeAssociation/norms_degrees.pickle'),
      'CCs': ps.load('data/FreeAssociation/norms_CCs.pickle'),
      'BCs': ps.load('data/FreeAssociation/norms_BCs.pickle'),
      'PR_scores': ps.load('data/FreeAssociation/norms_PR_scores.pickle')}
```

## Plot overlap of high index nodes

```python
# Imports
from __future__ import division

from operator import itemgetter

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot
%matplotlib inline
```

```python
# For each feature type, plot the cumulated overlap
fig, axs = mpl.pyplot.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
for j, ((wn_ftype, wn_fvalues), (fa_ftype, fa_fvalues)) in enumerate(zip(wn.iteritems(), fa.iteritems())):
    common_words = set(wn_fvalues.keys()).intersection(set(fa_fvalues.keys()))
    
    wn_sorted_tuples = sorted([wf for wf in wn_fvalues.iteritems() if wf[0] in common_words],
                              key=itemgetter(1), reverse=True)
    fa_sorted_tuples = sorted([wf for wf in fa_fvalues.iteritems() if wf[0] in common_words],
                              key=itemgetter(1), reverse=True)
    
    xs = np.arange(10, 5000, dtype=int)
    top_overlaps = np.zeros_like(xs, dtype=float)
    for i, ntop in enumerate(xs):
        wn_top = set(wf[0] for wf in wn_sorted_tuples[:ntop])
        fa_top = set(wf[0] for wf in fa_sorted_tuples[:ntop])
        top_overlaps[i] = len(wn_top.intersection(fa_top)) / np.min([ntop, len(wn_top)])
    
    axs.ravel()[j].plot(xs, top_overlaps, label=wn_ftype.replace('_', ' '))
    axs.ravel()[j].legend(loc='best')

fig.tight_layout()
fig.text(0.5, 1.03, 'Overlap percentages between FA and WN (restricted to common words)', ha='center', fontsize=15)
fig.text(- 0.03, 0.5, 'Overlap percentage', rotation='vertical', ha='center', va='center', fontsize=15)
fig.text(0.5, - 0.03, 'Number of words from top', ha='center', fontsize=15)
```

## Distance between ranks (normalized)

```python
# For each feature type, plot distribution of rank distances
fig1, axs1 = mpl.pyplot.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
fig2, axs2 = mpl.pyplot.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
for j, ((wn_ftype, wn_fvalues), (fa_ftype, fa_fvalues)) in enumerate(zip(wn.iteritems(), fa.iteritems())):
    common_words = set(wn_fvalues.keys()).intersection(set(fa_fvalues.keys()))
    
    wn_sorted_tuples = sorted([wf for wf in wn_fvalues.iteritems() if wf[0] in common_words],
                              key=itemgetter(1), reverse=True)
    fa_sorted_tuples = sorted([wf for wf in fa_fvalues.iteritems() if wf[0] in common_words],
                              key=itemgetter(1), reverse=True)
    wn_ranks = dict((wf[0], i) for i, wf in enumerate(wn_sorted_tuples))
    fa_ranks = dict((wf[0], i) for i, wf in enumerate(fa_sorted_tuples))
    
    ranks_array = np.array([[wn_ranks[w], fa_ranks[w]] for w in common_words])
    ranks_mins = np.min(ranks_array, 1)
    ranks_maxs = np.max(ranks_array, 1)
    ranks_signs = np.sign(ranks_array[:, 0] - ranks_array[:, 1])
    ranks_distances = ranks_maxs - ranks_mins
    ranks_distances_log = np.log(ranks_maxs + 1) - np.log(ranks_mins + 1)
    
    axs1.ravel()[j].hist(ranks_distances, label=wn_ftype.replace('_', ' '), bins=50, alpha=0.5)
    axs1.ravel()[j].legend(loc='best')
    
    axs2.ravel()[j].hist(ranks_distances_log, label=wn_ftype.replace('_', ' '), bins=50, alpha=0.5)
    axs2.ravel()[j].legend(loc='best')

fig1.tight_layout()
fig1.text(0.5, 1.03, 'Differences in ranks between FA and WN (for common words)', ha='center', fontsize=15)
fig1.text(- 0.03, 0.5, 'Distribution of rank difference', rotation='vertical', ha='center', va='center', fontsize=15)
fig1.text(0.5, - 0.03, '$|r_{WN} - r_{FA}|$', ha='center', fontsize=15)

fig2.tight_layout()
fig2.text(0.5, 1.03, 'Differences in log-ranks between FA and WN (for common words)', ha='center', fontsize=15)
fig2.text(- 0.03, 0.5, 'Distribution of log-rank difference', rotation='vertical', ha='center', va='center', fontsize=15)
fig2.text(0.5, - 0.03, '$|\log(r_{WN}) - \log(r_{FA})|$', ha='center', fontsize=15)
```
