# How does the distribution of featuresn evolve due to repeated effect of substitutions?

## Load and prep

```python
# Duration of a timebag, in days
timebag_size = 1.0
# Type of exclusion from previous timebags
exclusion_type = 'no'
# Minimum number of days a cluster must have to be included in the analysis
min_days = 20
# Maximum number of days a cluster must have to be included in the analysis
max_days = 80
# Clusters filtering type to use
filtering_type = 'filtered'

file_postfix = ('_F{filtering_type}_R{min_days}-{max_days}_D{timebag_size}_'
                'E{exclusion_type}').format(filtering_type=filtering_type,
                                            min_days=min_days, max_days=max_days,
                                            timebag_size=timebag_size, exclusion_type=exclusion_type)
```

```python
# First some imports
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

import settings as st
from linguistics.treetagger import get_tagger
from linguistics.wn import lemmatize as relemmatize
from linguistics.distance import levenshtein
from datainterface.redistools import RedisReader
import datainterface.picklesaver as ps
from util.generic import ProgressInfo
tagger = get_tagger()
```

```python
# Load the clusters, the features, and initialize the results
clusters = RedisReader(st.__getattribute__('redis_mt_clusters_{}_pref'.format(filtering_type)))
features = dict((fsrc, dict((ftype, ps.load(fdetails['file'].format('all')))
                            for ftype, fdetails in fsrc_details.iteritems()))
                for fsrc, fsrc_details in st.mt_analysis_features.iteritems())
macro_features = dict((fsrc, dict((ftype, [])
                                  for ftype in fsrc_values.iterkeys()))
                      for fsrc, fsrc_values in features.iteritems())
```

```python
from linguistics.distance import hamming_word
    
def is_in_sphere(ref_strings, string):
    """Whether or not `string` is in the hamming_word-sphere at
    distance 1 surrounding the `refstrings`."""
    
    # Lemmatize BEFORE hamming
    string_lems = [relemmatize(lem) for lem in tagger.Lemmatize(string)]
    for ref_string in ref_strings:
        ref_string_lems = [relemmatize(ref_lem) for ref_lem in tagger.Lemmatize(ref_string)]
        
        # Hamming distance (=subs detect)
        if len(string_lems) != len(ref_string_lems):
            continue
        diff = np.where([w1 != w2 for w1, w2 in zip(ref_string_lems, string_lems)])[0]
        if len(diff) != 1:
            continue
        
        # Check this difference is not trivial (=test_real)
        l1, l2 = ref_string_lems[diff], string_lems[diff]
        if levenshtein(l1, l2) > 1:
            return True
    return False


def get_filtered_relems_and_strings(tbg, ref_strings, excl_strings):
    """List the relemmatized words (once per occurrence) appearing in `tbg`,
    and at distance one from `ref_strings`, excluding strings in `excl_strings`.
    
    Quotes are first lemmatized with the TreeTagger lemmatizer, then
    relemmatized as in the mining analysis, with the WordNet morphy
    method.
    
    Parameters
    ----------
    tbg : TimeBag
        The timebag to work on.
    ref_strings : list of strings
        The reference string from which to compute distance; if None, then ignored.
    excl_strings : list of strings
        The reference string from which to exclude; if None, then ignored.
    
    Returns
    -------
    list
        The list of relemmatized words.
    set
        The set of new strings considered to the product of this iteration.
    
    """
    
    filtered_relems = []
    new_strings = set([])

    for string, freq in zip(tbg.strings, tbg.tot_freqs):
        # If our current string is in the set of banned strings,
        # jump to next one
        if excl_strings is not None and string in excl_strings:
            continue
        
        # If our current string is not in the sphere around ref_strings,
        # jump to next one
        if ref_strings is not None and not is_in_sphere(ref_strings, string):
            continue

        new_strings.add(string)
        relems = [relemmatize(lem) for lem in tagger.Lemmatize(string)]
        filtered_relems.extend(relems * freq)

    return filtered_relems, new_strings
```

## Mining

```python
class Excluder(object):
    
    def __init__(self):
        self.prevtbg = None
        self.prevall_strings = None

    def _exclusion_no(self):
        return None
    
    def _exclusion_prevall(self):
        if self.prevall_strings is not None:
            # Copy since we modify the prevall_strings in _update
            return self.prevall_strings.copy()
        else:
            return None
    
    def _exclusion_prevtbg(self):
        if self.prevtbg is not None:
            return self.prevtbg.strings
        else:
            return None
    
    def _update(self, tbg):
        self.prevtbg = tbg
        try:
            self.prevall_strings.update(tbg.strings)
        except AttributeError:
            self.prevall_strings = set(tbg.strings)

    def __call__(self, tbg):
        ret = self.__getattribute__('_exclusion_' + exclusion_type)()
        self._update(tbg)
        return ret
```

```python
try:
    #raise IOError
    # See if we already computed this
    macro_features = ps.load('data/MemeTracker/macro/macro_features{}.pickle'.format(file_postfix))
    clusters_log = ps.load('data/MemeTracker/macro/clusters_log{}.pickle'.format(file_postfix))
    print 'Loaded from pickle.'
except IOError:
    n_clusters_kept = 0
    clusters_log = []
    progress = ProgressInfo(len(clusters), 10, 'clusters')
    
    for c in clusters.itervalues():
        progress.next_step()
        
        # Skip if out of min-max
        c.build_timeline()
        if c.timeline.span_days > max_days or c.timeline.span_days < min_days:
            continue
        n_clusters_kept += 1
        
        cluster_log = (c, [])
        should_log = False
        excluder = Excluder()
        tbgs = c.build_timebags(timebag_size)
        for i, tbg in enumerate(tbgs):
            # Substitutions
            filtered_relems, new_strings = get_filtered_relems_and_strings(tbg, None if i == 0 else new_strings,
                                                                           excluder(tbg))
            
            # Features
            for fsrc, fsrc_values in features.iteritems():
                for ftype, feature in fsrc_values.iteritems():
            
                    # Get the feature values for the extracted words
                    tbg_macro_features = [feature[relem] for relem in filtered_relems if relem in feature]
                    try:
                        # Either add it to the existing list
                        macro_features[fsrc][ftype][i].extend(tbg_macro_features)
                    except IndexError:
                        # Or create the list for this timebag if necessary
                        macro_features[fsrc][ftype].append(tbg_macro_features)
                        # Then check that the item we just created was indeed for the right timebag.
                        macro_features[fsrc][ftype][i]

            # Logging
            if i > 0 and len(new_strings) > 0:
                should_log = True
            cluster_log[1].append(([q for q in c.quotes.itervalues() if q.string in tbg.strings],
                                   [q for q in c.quotes.itervalues() if q.string in new_strings]))
        
        # Logging
        if should_log:
            clusters_log.append(cluster_log)
    
    # Convert to numpy arrays
    print 'Kept {} clusters from days filter ({}-{}).'.format(n_clusters_kept, min_days, max_days)
    print '{} clusters have at least one occurrence considered new (logged).'.format(len(clusters_log))
    for fsrc, fsrc_values in features.iteritems():
        for ftype in fsrc_values.iterkeys():
            for i, macro_features_tbg in enumerate(macro_features[fsrc][ftype]):
                macro_features[fsrc][ftype][i] = np.array(macro_features_tbg)
    
    # Save to disk
    ps.save(macro_features, 'data/MemeTracker/macro/macro_features{}.pickle'.format(file_postfix))
    ps.save(clusters_log, 'data/MemeTracker/macro/clusters_log{}.pickle'.format(file_postfix))
```

## Meta-analysis

### Until which timebag clusters keep having new quotes

```python
last_novel_tbg_per_cluster = np.zeros(len(clusters_log), dtype=int)
for i, (c, tbgs_log) in enumerate(clusters_log):
    last_novel_tbg_per_cluster[i] = np.array([len(tbg_log[1]) for tbg_log in tbgs_log]).nonzero()[0][-1]

last_novel_tbg = last_novel_tbg_per_cluster.max()
bins = np.linspace(- 0.5, last_novel_tbg + 0.5, last_novel_tbg + 2)
plt.hist(last_novel_tbg_per_cluster, bins=bins, log=True)
```

### Histograms of number of new quotes per timebag

```python
n_new_per_tbg = []

# Collect values
for c, tbgs_log in clusters_log:
    for i, tbg_log in enumerate(tbgs_log):
        try:
            # Either add it to the existing list
            n_new_per_tbg[i].append(len(tbg_log[1]))
        except IndexError:
            # Or create the list for this timebag if necessary
            n_new_per_tbg.append([len(tbg_log[1])])
            # Then check that the item we just created was indeed for the right timebag.
            n_new_per_tbg[i]

# Convert to numpy arrays
np_n_new_per_tbg = []
for i, n_new_this_tbg in enumerate(n_new_per_tbg):
    np_n_new_per_tbg.append(np.array(n_new_this_tbg))
max_new_in_tbg = max([np_n_new_this_tbg.max() for np_n_new_this_tbg in np_n_new_per_tbg])

# Plot
bins = np.linspace(- 0.5, max_new_in_tbg + 0.5, max_new_in_tbg + 2)
fig, axs = plt.subplots(1, last_novel_tbg + 1, figsize=(3 * (last_novel_tbg - 1), 2), sharey=True)
for i in range(last_novel_tbg + 1):
    axs[i].hist(np_n_new_per_tbg[i], bins=bins, log=True)
    axs[i].set_xlim(bins[0], bins[-1])
    axs[i].set_xlabel('Bag {}'.format(i))
```

### Detailed log

```python
log_range = np.where(last_novel_tbg_per_cluster < 5)[0][:10]
```

```python
def plot_cluster_novelty(c, tbgs_log):
    blue = sb.color_palette()[5]
    red = sb.color_palette()[2]
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 3))
    ax.set_title('Cluster #{}, {} quotes, total freq {}'.format(c.id, c.n_quotes, c.tot_freq))
    ax.set_xlim(- 0.5, last_novel_tbg - 0.5)
    ax.set_xticks(range(last_novel_tbg))
    ax.set_xlabel('Bag number')
    ax.set_ylim(- 0.5, c.n_quotes - 0.5)
    ax.set_yticks(range(c.n_quotes))
    ax.set_ylabel('Quote occurrences')
    
    qt_coords = dict((q.id, i) for i, q in enumerate(c.quotes.itervalues()))
    for i, tbg_log in enumerate(tbgs_log[:last_novel_tbg]):
        tbg_qt_coords = np.array([qt_coords[q.id] for q in tbg_log[0]])
        tbg_new_qt_coords = np.array([qt_coords[q.id] for q in tbg_log[1]])
        ax.plot(np.ones_like(tbg_qt_coords) * i, tbg_qt_coords, 'o', color=blue, markersize=8)
        ax.plot(np.ones_like(tbg_new_qt_coords) * i, tbg_new_qt_coords, 'o', color=red, markersize=12)
    fig.show()


#plot_cluster_novelty(print_clusters[1][0], print_clusters[1][1])
for i in log_range:
    c, tbgs_log = clusters_log[i]
    plot_cluster_novelty(c, tbgs_log)
```
