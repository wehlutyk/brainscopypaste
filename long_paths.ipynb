```python
# Imports
import networkx as nx
import numpy as np
from nltk.corpus import wordnet as wn

import datainterface.picklesaver as ps
from analyze.paths import PathsAnalysis
```

```python
# Load substitution data
subs = ps.load('../data/MemeTracker/mining/Ffiltered_Mtbgs_Sno_Pall_N5_substitutions.pickle')

# Start paths analyzer
pa = PathsAnalysis(None, subs)
pa.load()
```

```python
# Compute susbtitution path lengths
lengths = []
for s in pa.data:
    try:
        lengths.append(NaN)
        l = nx.shortest_path_length(pa.G, s.lem1, s.lem2)
        lengths[-1] = l
    except (nx.NetworkXError, nx.NetworkXNoPath):
        continue
lengths = np.array(lengths)

# Find long paths
long_def = 7
long_paths = np.where(lengths >= long_def)[0]
print 'Found {} long paths (>= {})'.format(len(long_paths), long_def)
```
