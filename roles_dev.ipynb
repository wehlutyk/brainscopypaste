# How do clusters compose the bulk of data in time?

```python
# Duration of a timebag, in days
timebag_size = 1.0
# Minimum number of days a cluster must have to be included in the analysis
min_days = 0
# Maximum number of days a cluster must have to be included in the analysis
max_days = 500 # no filter
# Clusters filtering type to use
filtering_type = 'full'

file_postfix = ('_F{filtering_type}_R{min_days}-{max_days}_'
                'D{timebag_size}').format(filtering_type=filtering_type,
                                          min_days=min_days, max_days=max_days,
                                          timebag_size=timebag_size)
```

```python
# First some imports
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
import pandas as pd

import settings as st
import datainterface.picklesaver as ps
from linguistics.treetagger import get_tagger
from linguistics.wn import lemmatize as relemmatize
from datainterface.redistools import RedisReader
from util.generic import ProgressInfo
tagger = get_tagger()
```

```python
# Load the clusters, the features, and initialize the results
pref = st.__getattribute__('redis_mt_clusters_{}_pref'.format(filtering_type)) if filtering_type != 'full' else st.redis_mt_clusters_pref
clusters = RedisReader(pref)
features = dict((fsrc, dict((ftype, ps.load(fdetails['file'].format('all')))
                            for ftype, fdetails in fsrc_details.iteritems()))
                for fsrc, fsrc_details in st.mt_analysis_features.iteritems())
feature_averages = dict((fsrc, dict((ftype, [])
                                    for ftype in fsrc_values.iterkeys()))
                        for fsrc, fsrc_values in features.iteritems())
n_quotes = []
n_occurrences = []
clspans = []
```

```python
def get_relems(tbg):
    #"""List the relemmatized words (once per quote) appearing in `tbg`."""
    """List the relemmatized words (once per occurrence) appearing in `tbg`."""
    
    tbg_relems = []
    
    for string, freq in zip(tbg.strings, tbg.tot_freqs):
        string_relems = [relemmatize(lem) for lem in tagger.Lemmatize(string)]
        tbg_relems.extend(string_relems * freq)

    return tbg_relems
```

```python
try:
    #raise IOError
    # See if we already computed this
    feature_averages = ps.load('data/MemeTracker/macro/roles_feature_averages{}.pickle'.format(file_postfix))
    n_quotes = ps.load('data/MemeTracker/macro/roles_n_quotes{}.pickle'.format(file_postfix))
    n_occurrences = ps.load('data/MemeTracker/macro/roles_n_occurrences{}.pickle'.format(file_postfix))
    clspans = ps.load('data/MemeTracker/macro/roles_clspans{}.pickle'.format(file_postfix))
    print 'Loaded from pickle.'
except IOError:
    n_clusters_kept = 0
    progress = ProgressInfo(len(clusters), 10, 'clusters')
    
    for c in clusters.itervalues():
        progress.next_step()
        
        # Skip if out of min-max
        c.build_timeline()
        if c.timeline.span_days > max_days or c.timeline.span_days < min_days:
            continue
        n_clusters_kept += 1
        
        tbgs = c.build_timebags(timebag_size)
        for i, tbg in enumerate(tbgs):
            # Relems
            relems = get_relems(tbg)
            
            # Feature averages
            for fsrc, fsrc_values in features.iteritems():
                for ftype, feature in fsrc_values.iteritems():
                    average = np.mean([feature[relem] for relem in relems if relem in feature])
                    if np.isnan(average):
                        continue
                    try:
                        # Either add it to the existing list
                        feature_averages[fsrc][ftype][i].append(average)
                    except IndexError:
                        # Or create the list for this timebag if necessary
                        for j in range(1 + i - len(feature_averages[fsrc][ftype])):
                            feature_averages[fsrc][ftype].append([])
                        # Then add to the new list for this timebag
                        feature_averages[fsrc][ftype][i].append(average)
            
            # Number of quotes
            try:
                n_quotes[i].append(len(tbg.strings))
            except IndexError:
                n_quotes.append([len(tbg.strings)])
                n_quotes[i]
            
            # Number of occurrences
            try:
                n_occurrences[i].append(tbg.tot_freq)
            except IndexError:
                n_occurrences.append([tbg.tot_freq])
                n_occurrences[i]
            
            # Cluster spans
            try:
                clspans[i].append(c.timeline.span_days)
            except IndexError:
                clspans.append([c.timeline.span_days])
                clspans[i]
    
    # Convert to numpy arrays
    print 'Kept {} clusters from days filter ({}-{}).'.format(n_clusters_kept, min_days, max_days)
    for fsrc, fsrc_values in features.iteritems():
        for ftype in fsrc_values.iterkeys():
            for i, feature_averages_tbg in enumerate(feature_averages[fsrc][ftype]):
                feature_averages[fsrc][ftype][i] = np.array(feature_averages_tbg)
    for i, n_quotes_tbg in enumerate(n_quotes):
        n_quotes[i] = np.array(n_quotes_tbg)
    for i, n_occurrences_tbg in enumerate(n_occurrences):
        n_occurrences[i] = np.array(n_occurrences_tbg)
    for i, clspans_tbg in enumerate(clspans):
        clspans[i] = np.array(clspans_tbg)
    
    # Save to disk
    ps.save(feature_averages, 'data/MemeTracker/macro/roles_feature_averages{}.pickle'.format(file_postfix))
    ps.save(n_quotes, 'data/MemeTracker/macro/roles_n_quotes{}.pickle'.format(file_postfix))
    ps.save(n_occurrences, 'data/MemeTracker/macro/roles_n_occurrences{}.pickle'.format(file_postfix))
    ps.save(clspans, 'data/MemeTracker/macro/roles_clspans{}.pickle'.format(file_postfix))
```

## Plot

```python
included_features = ['aoa_Kuperman',
                     'mt_frequencies',
                     'fa_CCs',
                     'cmu_MNphonemes',
                     'wn_MNSyns']

# Translating internal feature names to public names
tr_map = {'aoa_Kuperman': 'Age of acquisition',
          'cmu_MNphonemes': r'$\left< \right.$ \# phonemes $\left. \right>$',
          'cmu_MNsyllables': r'$\left< \right.$ \# syllables $\left. \right>$',
          'fa_BCs': 'FA log-betweenness',
          'fa_CCs': 'FA log-clustering',
          'fa_degrees': 'FA log-degree',
          'fa_PR_scores': 'FA log-pagerank',
          'mt_frequencies': 'Log-frequency',
          'mt_start_frequencies': 'Log-frequency in start quotes',
          'wn_BCs': 'WN log-betweenness',
          'wn_CCs': 'WN log-clustering',
          'wn_degrees': 'WN log-degree',
          'wn_MNSyns': r'Log $\left< \right.$ \# synonyms $\left. \right>$',
          'wn_NSigns': r'Log \# meanings',
          'wn_PR_scores': 'WN log-pagerank'}

def translate(labels):
    """Translate a list of internal labels to human-readable ones."""
    if isinstance(labels, list):
        return [tr_map[l] for l in labels]
    else:
        return tr_map[labels]
```

```python
from __future__ import division

nbins = 15             # Number of bins to use in histograms
n_timebags_shown = 10  # Number of timebags to plot for each feature
n_timebags_step = 15    # Step between two plotted timebags
small_fontsize = 8     # fontsize for small, not so important text
med_fontsize = 10      # fontsize for medium, moderately important text
big_fontsize = 14      # fontsize for big, important text
palette = sb.color_palette('husl', len(included_features))

fig, axs = mpl.pyplot.subplots(len(included_features), n_timebags_shown,
                               figsize=(1.8 * n_timebags_shown, 2 * len(included_features)),
                               sharey='row')

mus = np.zeros((len(included_features), n_timebags_shown))
sigma2s = np.zeros_like(mus)
allmus = np.zeros((len(included_features), n_timebags_shown * n_timebags_step))
allsigma2s = np.zeros_like(allmus)

for i, ffullname in enumerate(included_features):
    ffullname_parts = ffullname.split('_')
    fsrc = ffullname_parts[0]
    ftype = '_'.join(ffullname_parts[1:])
    
    log = st.mt_analysis_features[fsrc][ftype]['log']
    feature_averages_feature = feature_averages[fsrc][ftype]
    
    vals = []
    for lvals in feature_averages_feature[:n_timebags_shown * n_timebags_step:n_timebags_step]:
        # Convert to log if asked to
        vals.append(np.log(lvals) if log else lvals)
    
    for j, lvals in enumerate(feature_averages_feature[:n_timebags_shown * n_timebags_step]):
        # Compute mu and sigma2
        llvals = np.log(lvals) if log else lvals
        allmus[i, j] = llvals.mean()
        allsigma2s[i, j] = llvals.var()
        
    vmin = np.array([v.min() for v in vals]).min()
    vmax = np.array([v.max() for v in vals]).max()
    bins = np.linspace(vmin, vmax, nbins)
    bin_width = bins[1] - bins[0]
    bin_lefts = bins[:-1]
    
    axs[i, 0].set_ylabel(translate(ffullname), fontsize=big_fontsize)
    for j in range(n_timebags_shown):
        # Compute mu and sigma2
        mus[i, j] = vals[j].mean()
        sigma2s[i, j] = vals[j].var()
        
        counts = np.histogram(vals[j], bins=bins)[0]
        probs = counts / counts.sum()
        axs[i, j].bar(bin_lefts, probs, bin_width, #log=log, 
                      color=palette[i], alpha=0.5)
        axs[i, j].set_xlim([bins.min(), bins.max()])
        axs[i, j].set_xticklabels(axs[i, j].get_xticks(), rotation=60, fontsize=small_fontsize)
        axs[i, j].set_yticklabels(axs[i, j].get_yticks(), fontsize=small_fontsize)

for j in range(n_timebags_shown):
    axs[-1, j].set_xlabel('bag {}'.format(j * n_timebags_step), fontsize=big_fontsize)

for i, ffullname in enumerate(included_features):
    ffullname_parts = ffullname.split('_')
    fsrc = ffullname_parts[0]
    ftype = '_'.join(ffullname_parts[1:])
    
    for j in range(n_timebags_shown):
        xmin, xmax = axs[i, j].get_xlim()
        xspan = xmax - xmin
        xpos = xmin + xspan * 0.95
        ymin, ymax = axs[i, j].get_ylim()
        yspan = ymax - ymin
        ypos = ymin + yspan * 0.95
        legend_text = r'$\mu = {:.3}, \sigma^2 = {:.3}$'.format(mus[i, j], sigma2s[i, j])
        axs[i, j].text(xpos, ypos, legend_text,
                       ha='right', va='top', fontsize=med_fontsize,
                       bbox=dict(facecolor=sb.desaturate('#EAEAF2', 0.2), alpha=0.3))

fig.tight_layout()
```

```python
fig, axs = mpl.pyplot.subplots(1, 3, figsize=(15, 4))
ax_color = '#999999'
axs[0].plot(range(n_timebags_shown * n_timebags_step), np.ones(n_timebags_shown * n_timebags_step) * 100, '--', color=ax_color)
axs[1].plot(range(n_timebags_shown * n_timebags_step), np.ones(n_timebags_shown * n_timebags_step) * 100, '--', color=ax_color)
axs[2].plot([100, 100], [0, 200], '--', color=ax_color)
axs[2].plot([0, 200], [100, 100], '--', color=ax_color)
for i, ffullname in enumerate(included_features):
    axs[0].plot(range(0, n_timebags_step * n_timebags_shown),
                100 * allmus[i] / allmus[i][0],
                color=palette[i], label=translate(ffullname))
    axs[1].plot(range(0, n_timebags_step * n_timebags_shown),
                100 * np.sqrt(allsigma2s[i] / allsigma2s[i][0]),
                color=palette[i], label=translate(ffullname))
    axs[2].plot(100 * allmus[i] / allmus[i][0], 100 * np.sqrt(allsigma2s[i] / allsigma2s[i][0]),
                color=palette[i], label=translate(ffullname))
    axs[2].plot(100 * allmus[i] / allmus[i][0], 100 * np.sqrt(allsigma2s[i] / allsigma2s[i][0]),
                color=palette[i], marker='o', markersize=6)

axs[0].legend(loc='best')
axs[0].set_xlabel('Bag number')
axs[0].set_ylabel('100-index')
axs[0].set_title(r'Evolution of $\mu$ w.r.t. first value')
ymin0, ymax0 = axs[0].get_ylim()

axs[1].legend(loc='best')
axs[1].set_xlabel('Bag number')
axs[1].set_ylabel('100-index')
axs[1].set_title(r'Evolution of $\sqrt{\sigma^2}$ w.r.t. first value')
ymin1, ymax1 = axs[1].get_ylim()

ymin = min([ymin0, ymin1])
ymax = max([ymax0, ymax1])
#axs[0].set_ylim(ymin, ymax)
#axs[1].set_ylim(ymin, ymax)

axs[2].legend(loc='best')
axs[2].set_xlabel(r'$\mu$')
axs[2].set_ylabel(r'$\sqrt{\sigma^2}$')
axs[2].set_title(r'$\sqrt{\sigma^2}$ as a function of $\mu$')
axs[2].set_xlim(ymin0, ymax0)
axs[2].set_ylim(ymin1, ymax1)

fig.tight_layout()
```

```python
def plot_array_tbgs(values, label, log=False):
    fig, axs = mpl.pyplot.subplots(1, n_timebags_shown,
                                   figsize=(2 * n_timebags_shown, 2),
                                   sharey='row')
    
    mus = np.zeros(n_timebags_shown)
    sigma2s = np.zeros_like(mus)
    allmus = np.zeros(n_timebags_shown * n_timebags_step)
    allsigma2s = np.zeros_like(allmus)
    
    vmin = np.array([v.min() for v in values]).min()
    vmax = np.array([v.max() for v in values]).max()
    bins = np.linspace(vmin, vmax, nbins)
    bin_width = bins[1] - bins[0]
    bin_lefts = bins[:-1]
    
    for j in range(n_timebags_shown * n_timebags_step):
        allmus[j] = values[j].mean()
        allsigma2s[j] = values[j].var()
    
    axs[0].set_ylabel(label, fontsize=big_fontsize)
    for j in range(n_timebags_shown):
        # Compute mu and sigma2
        mus[j] = values[j * n_timebags_step].mean()
        sigma2s[j] = values[j * n_timebags_step].var()
        
        counts = np.histogram(values[j * n_timebags_step], bins=bins)[0]
        probs = counts / counts.sum()
        axs[j].bar(bin_lefts, probs, bin_width, alpha=0.5, log=log)
        axs[j].set_xlim([bins.min(), bins.max()])
        axs[j].set_xticklabels(axs[j].get_xticks(), rotation=60, fontsize=small_fontsize)
        axs[j].set_yticklabels(axs[j].get_yticks(), fontsize=small_fontsize)
        axs[j].set_xlabel('bag {}'.format(j * n_timebags_step), fontsize=big_fontsize)

    for j in range(n_timebags_shown):
        xmin, xmax = axs[j].get_xlim()
        xspan = xmax - xmin
        xpos = xmin + xspan * 0.95
        ymin, ymax = axs[j].get_ylim()
        yspan = ymax - ymin
        ypos = ymin + yspan * 0.95
        legend_text = r'$\mu = {:.3}, \sigma^2 = {:.3}$'.format(mus[j], sigma2s[j])
        axs[j].text(xpos, ypos, legend_text,
                    ha='right', va='top', fontsize=med_fontsize,
                    bbox=dict(facecolor=sb.desaturate('#EAEAF2', 0.2), alpha=0.3))

    fig.tight_layout()
    
    # ----------------------
    
    fig, axs = mpl.pyplot.subplots(1, 3, figsize=(15, 4))
    ax_color = '#999999'
    axs[0].plot(range(n_timebags_shown * n_timebags_step), np.ones(n_timebags_shown * n_timebags_step) * 100, '--', color=ax_color)
    axs[1].plot(range(n_timebags_shown * n_timebags_step), np.ones(n_timebags_shown * n_timebags_step) * 100, '--', color=ax_color)
    axs[2].plot([100, 100], [0, 200], '--', color=ax_color)
    axs[2].plot([0, 200], [100, 100], '--', color=ax_color)
    axs[0].plot(range(0, n_timebags_step * n_timebags_shown),
                100 * allmus / allmus[0], label=label)
    axs[1].plot(range(0, n_timebags_step * n_timebags_shown),
                100 * np.sqrt(allsigma2s / allsigma2s[0]), label=label)
    axs[2].plot(100 * allmus / allmus[0], 100 * np.sqrt(allsigma2s / allsigma2s[0]),
                label=label)
    axs[2].plot(100 * allmus / allmus[0], 100 * np.sqrt(allsigma2s / allsigma2s[0]),
                marker='o', markersize=6)

    axs[0].legend(loc='best')
    axs[0].set_xlabel('Bag number')
    axs[0].set_ylabel('100-index')
    axs[0].set_title(r'Evolution of $\mu$ w.r.t. first value')
    ymin0, ymax0 = axs[0].get_ylim()
    
    axs[1].legend(loc='best')
    axs[1].set_xlabel('Bag number')
    axs[1].set_ylabel('100-index')
    axs[1].set_title(r'Evolution of $\sqrt{\sigma^2}$ w.r.t. first value')
    ymin1, ymax1 = axs[1].get_ylim()
    
    ymin = min([ymin0, ymin1])
    ymax = max([ymax0, ymax1])
    #axs[0].set_ylim(ymin, ymax)
    #axs[1].set_ylim(ymin, ymax)
    
    axs[2].legend(loc='best')
    axs[2].set_xlabel(r'$\mu$')
    axs[2].set_ylabel(r'$\sqrt{\sigma^2}$')
    axs[2].set_title(r'$\sqrt{\sigma^2}$ as a function of $\mu$')
    axs[2].set_xlim(ymin0, ymax0)
    axs[2].set_ylim(ymin1, ymax1)
    
    fig.tight_layout()
```

```python
plot_array_tbgs(n_quotes, 'n quotes', log=True)
```
