# How do features vary on substitution? What susceptibility to substitution do they induce?

## Data-loading and formatting

See below for graphs.

```python
# Initial imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

from analyze.args import AnalysisArgs
from analyze.substitutions import SubstitutionsAnalyzer
```

```python
included_features = ['aoa_Kuperman',
                     'mt_frequencies',
                     'fa_CCs',
                     'fa_PR_scores',
                     'cmu_MNphonemes',
                     'wn_NSigns']

# Build the analysis classes
aa = AnalysisArgs({'ff': 'filtered',
                   'model': 'tbgs',
                   'POS': 'all',
                   'substrings': False,
                   'timebag_size': 2.0,
                   'features': included_features,
                   'positions': False,
                   'paths': False,
                   'save': False,
                   'overwrite': False,
                   'show': True})
sa = SubstitutionsAnalyzer(aa)

print 'Loaded the {} substitutions that have been mined.'.format(len(sa.substitutions))
```

## Susceptibility to substitution for all the features

*feature value* ~ *susceptibility value*

**All plots are with zero values removed.** This consists in looking at the susceptibilities *for words that have been substituted at least once*.

```python
# Translating internal feature names to public names
tr_map = {'aoa_Kuperman': 'Age of acquisition',
          'cmu_MNphonemes': r'$\left< \right.$ \# phonemes $\left. \right>$',
          'cmu_MNsyllables': r'$\left< \right.$ \# syllables $\left. \right>$',
          'fa_BCs': 'FA log-betweenness',
          'fa_CCs': 'FA log-clustering',
          'fa_degrees': 'FA log-degree',
          'fa_PR_scores': 'FA log-pagerank',
          'mt_frequencies': 'Log-frequency',
          'mt_start_frequencies': 'Log-frequency in start quotes',
          'wn_BCs': 'WN log-betweenness',
          'wn_CCs': 'WN log-clustering',
          'wn_degrees': 'WN log-degree',
          'wn_MNSyns': r'Log $\left< \right.$ \# synonyms $\left. \right>$',
          'wn_NSigns': r'Log \# meanings',
          'wn_PR_scores': 'WN log-pagerank'}

def translate(labels):
    """Translate a list of internal labels to human-readable ones."""
    if isinstance(labels, list):
        return [tr_map[l] for l in labels]
    else:
        return tr_map[labels]
```

```python
# Build susceptibilities and filter out zeros
for a in sa.analyses:
    try:
        a.f_susceptibilities
    except AttributeError:
        a.build_susceptibilities()
        # The try-except here lets us run this cell several times without overwriting
        # the original unfiltered susceptibilities each time.
        a.f_susceptibilities_orig = a.f_susceptibilities
    
    a.f_susceptibilities = a.f_susceptibilities_orig[np.nonzero(a.f_susceptibilities_orig[:, 1])]

# Set names
for a in sa.analyses:
    a.feature.publicname = translate(a.feature.fullname.replace(' ', '_').rstrip('_all'))

# Convert to DataFrame
sdata = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.f_susceptibilities) for a in sa.analyses]),
                      'feature value': np.concatenate([a.f_susceptibilities[:, 0] for a in sa.analyses]),
                      'substitution susceptibility': np.concatenate([a.f_susceptibilities[:, 1] for a in sa.analyses])})
```

```python
# Our plotting function for a given feature
def plot_susceptibility(xvalues, svalues, **kwargs):
    color = kwargs.get('color', 'blue')
    
    bins = np.linspace(xvalues.iloc[0], xvalues.iloc[-1], 20)
    xs = (bins[1:] + bins[:-1]) / 2
    ys = np.zeros_like(xs)
    cs = np.zeros_like(xs)
    for i in range(len(bins) - 1):
        idx = np.where((bins[i] <= xvalues) * (xvalues < bins[i + 1]))[0]
        yss = svalues.iloc[idx]
        ys[i] = yss.mean()
        cs[i] = 1.96 * yss.std() / np.sqrt(len(idx) - 1)
    
    plt.hlines(0, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.hlines(1, xs[0], xs[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.plot(xs, ys, '-', lw=2, color=color)
    plt.fill_between(xs, ys - cs, ys + cs, color=sb.desaturate(color, 0.2), alpha=0.2)
    plt.xlim(xs[0], xs[-1])

# Plot!
sg = sb.FacetGrid(sdata, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1.5, size=3, sharex=False)
sg.map(plot_susceptibility, 'feature value', 'substitution susceptibility')
sg.set_titles('{col_name}')
```

## Feature variations on substitution

The first 4 rows are the network properties (betweenness, clustering, pagerank, and degrees), computed on the Free Association network (graphs to the left) and the WordNet network (graphs to the right).

The last 4 rows are the remaining features ("usual suspects").

```python
# Build variations
for a in sa.analyses:
    a.build_h0()
    a.build_variations()

# Convert to DataFrame
vdata_binned = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.bin_middles) for a in sa.analyses]),
                             'disappearing bin middles': np.concatenate([a.bin_middles for a in sa.analyses]),
                             'appearing bin averages': np.concatenate([a.daughter_d - a.feature.values.mean() for a in sa.analyses]),
                             'ci': np.concatenate([a.daughter_d_ci for a in sa.analyses])})
vdata_raw = pd.DataFrame({'feature': np.concatenate([[a.feature.publicname] * len(a.l2_f_mothers) for a in sa.analyses]),
                          'disappearing raw': np.concatenate([a.l2_f_mothers for a in sa.analyses]),
                          'appearing raw': np.concatenate([a.l2_f_daughters - a.feature.values.mean() for a in sa.analyses])})
```

```python
def plot_variations(xvalues, avalues, cis, **kwargs):
    color = kwargs.get('color', 'blue')
    
    plt.plot(xvalues, avalues, '-', lw=2, color=color)
    plt.hlines(0, xvalues.iloc[0], xvalues.iloc[-1], linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.fill_between(xvalues.values, avalues.values - cis.values, avalues.values + cis.values,
                     color=sb.desaturate(color, 0.2), alpha=0.2)
    plt.xlim(xvalues.iloc[0], xvalues.iloc[-1])
    # Rescale on Y if we're touching 0, making sure the horizontal line at y=0 is visible
    ymin, ymax = plt.ylim()
    if ymin == 0:
        ymin -= (ymax - ymin) / 10
    elif ymax == 0:
        ymax += (ymax - ymin) / 10
    plt.ylim(ymin, ymax)

# Plot!
vg = sb.FacetGrid(vdata_binned, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1.5, size=3, sharex=False, sharey=False)
vg.map(plot_variations, 'disappearing bin middles', 'appearing bin averages', 'ci')
vg.set_titles('{col_name}')
vg.set_xlabels(r'$f \left( \right.$ disappearing word $\left. \right)$')
vg.set_ylabels(r'$f \left( \right.$ appearing word $\left. \right) - \left< f \right>$')
```

```python
def kdeplot(xvalues, yvalues, **kwargs):
    color = kwargs.get('color', 'blue')
    sb.kdeplot(pd.DataFrame({'X': xvalues, 'Y': yvalues}), shade=True)
    xmin, xmax = plt.xlim()
    plt.hlines(0, xmin, xmax, linestyles='dashed', colors=sb.desaturate(color, 0.2))
    plt.xlim(xmin, xmax)

vg = sb.FacetGrid(vdata_raw, col='feature', col_wrap=3, col_order=translate(included_features),
                  aspect=1, size=4, sharex=False, sharey=False)
vg.map(kdeplot, 'disappearing raw', 'appearing raw')
vg.set_titles('{col_name}')
vg.set_xlabels(r'$f \left( \right.$ disappearing word $\left. \right)$')
vg.set_ylabels(r'$f \left( \right.$ appearing word $\left. \right) - \left< f \right>$')
```
