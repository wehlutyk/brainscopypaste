# How do features vary on substitution? What susceptibility to substitution do they induce?

## Data-loading and formatting

See below for graphs.

```python
# Initial imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from analyze.args import AnalysisArgs
from analyze.substitutions import SubstitutionsAnalyzer
```

```python
# Build the analysis classes
aa = AnalysisArgs({'ff': 'filtered',
                   'model': 'tbgs',
                   'POS': 'all',
                   'substrings': False,
                   'timebag_size': 5.0,
                   'features': ['cmu_MNphonemes',
                                #'cmu_MNsyllables',
                                'aoa',
                                'fa', 
                                'wn',
                                'mt'],
                   'positions': False,
                   'paths': False,
                   'save': False,
                   'overwrite': False,
                   'show': True})
sa = SubstitutionsAnalyzer(aa)
```

```python
# How many substitutions do we have?
len(sa.substitutions)
```

```python
# Order and group the analyses
from operator import itemgetter

net_ftypes = set(['BCs', 'CCs', 'PR_scores', 'degrees'])
sorted_analyses = sorted(sa.analyses, key=lambda a: (a.feature.data_src, a.feature.ftype))

fa_net_analyses_idx = np.where(map(lambda a: a.feature.data_src == 'fa' and a.feature.ftype in net_ftypes, sorted_analyses))[0]
fa_net_analyses = [sorted_analyses[i] for i in fa_net_analyses_idx]
wn_net_analyses_idx = np.where(map(lambda a: a.feature.data_src == 'wn' and a.feature.ftype in net_ftypes, sorted_analyses))[0]
wn_net_analyses = [sorted_analyses[i] for i in wn_net_analyses_idx]
rest_analyses_idx = set(range(len(sorted_analyses))).difference(set(fa_net_analyses_idx).union(set(wn_net_analyses_idx)))
rest_analyses = [sorted_analyses[i] for i in rest_analyses_idx]
```

## Feature variations on substitution

The first 4 rows are the network properties (betweenness, clustering, pagerank, and degrees), computed on the Free Association network (graphs to the left) and the WordNet network (graphs to the right).

The last 4 rows are the remaining features ("usual suspects").

```python
# Plot some analyses, grouped
for fa_a, wn_a in zip(fa_net_analyses, wn_net_analyses):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    fa_a.plot_variations(axs[0])
    wn_a.plot_variations(axs[1])
    fig.text(0.5, 0.95, fa_a.latexize(fa_a.feature.fullname + ' --- ' +
                                      fa_a.aa.title() + ' --- ' +
                                      wn_a.feature.fullname),
             ha='center')

for rest_a in rest_analyses:
    fig, ax = plt.subplots(figsize=(6, 4))
    rest_a.plot_variations(ax)
    fig.text(0.5, 0.95, rest_a.latexize(rest_a.aa.title() + ' --- ' +
                                        rest_a.feature.fullname),
             ha='center')
```

## Susceptibility to substitution for all the features

On the right, the scatter plots of *feature value* v. *susceptibility value* (each point represents a word). On the left, the average of those into 20 bins, with 95% confidence intervals in grey.

Not much to be said: there's loads happening on the right, but it doesn't show up on the left. **How can we visualize this better?** Apart from smarter binning (how?). Remove all zeros and ones (justification?)?

```python
# Plot susceptibilities
for a in sorted_analyses:
    fig, axs = plt.subplots(1, 2, sharey=True, figsize=(12, 4))
    
    # The whole scatter plot of susceptibilities
    a.build_susceptibilities()
    axs[1].scatter(a.f_susceptibilities[:, 0], a.f_susceptibilities[:, 1])
    
    # Averages in bins
    bins = np.linspace(a.f_susceptibilities[0, 0], a.f_susceptibilities[-1, 0], 20)
    xs = (bins[1:] + bins[:-1]) / 2
    ys = np.zeros_like(xs)
    cs = np.zeros_like(xs)
    for i in range(len(bins) - 1):
        idx = np.where((bins[i] <= a.f_susceptibilities[:, 0]) * (a.f_susceptibilities[:, 0] < bins[i + 1]))[0]
        yss = a.f_susceptibilities[idx, 1]
        ys[i] = yss.mean()
        cs[i] = 1.96 * yss.std() / np.sqrt(len(idx) - 1)
    
    axs[0].plot(xs, ys, 'b-', lw=2)
    axs[0].fill_between(xs, ys - cs, ys + cs, color='grey', alpha=0.2)
    a.print_fig_text(fig, a.aa.title())
```
